```python
from matplotlib.image import imread

image_source = imread('c:/data/images/ladybug.png')
image_source.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_source)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.cluster import KMeans
X = image_source.reshape(-1, 3) # (533, 800, 3) => (-1, 3) color  reshape
images = []
for k in range(2,9):
    kmeans = KMeans(n_clusters=k, random_state=10, n_init='auto').fit(X)
    img = kmeans.cluster_centers_[kmeans.labels_]
    img = img.reshape(image_source.shape)
    images.append(img)
```

```python
import cv2

plt.figure(figsize=(20,7))
for idx, img in enumerate(images):
    plt.subplot(2, 4, idx+1)
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(f'k={idx+2}')
    plt.axis("off")

plt.subplot(2,4,8)
plt.imshow(image_source)
plt.title('original image')
plt.axis("off")
plt.show()

#      color   
```
